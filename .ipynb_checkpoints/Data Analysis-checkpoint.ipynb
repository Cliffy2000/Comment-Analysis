```python
import math
import json
import time
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from tqdm.notebook import trange, tqdm
from matplotlib.patches import Patch
from matplotlib.font_manager import FontProperties
```

```python
cf = FontProperties(fname='AlibabaPuHuiTi-2-55-Regular.ttf')
```

```python
# paddle.enable_static()
# jieba.enable_paddle()
```

```python
totalData = []
for i in range(53):
    with open(f"DataFile{i+1}.json", 'r', encoding='utf-8') as tempFile:
        temp = json.load(tempFile)
        totalData += temp
```

```python
dataFrame = pd.DataFrame(totalData)
```

```python
dfUniqueUser = dataFrame.drop_duplicates(subset="userId", keep="first").copy()
nameCounts = dataFrame["userId"].value_counts()
dfUniqueUser["frequency"] = dfUniqueUser["userId"].map(nameCounts)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="userLevel", data=dfUniqueUser, order=[0,1,2,3,4,5,6], color="tab:red", edgecolor="white", linewidth=0.75, width=1)
sns.countplot(x="userLevel", data=dfUniqueUser[dfUniqueUser["frequency"] < 5], order=[0,1,2,3,4,5,6], color="tab:orange", edgecolor="white", linewidth=0.75, width=1)
sns.countplot(x="userLevel", data=dfUniqueUser[dfUniqueUser["frequency"] == 1], order=[0,1,2,3,4,5,6], color="tab:blue", edgecolor="white", linewidth=0.75, width=1)
plt.title("BA 用户等级分布", fontproperties=cf)
plt.xlabel("用户等级", fontproperties=cf)
plt.ylabel("频率", fontproperties=cf)
plt.margins(x=0.05, y=0.1)
legendPatches = [
    Patch(color="tab:red", label=">=5"), 
    Patch(color="tab:orange", label=">1, <5"), 
    Patch(color="tab:blue", label="= 1")
]
plt.legend(handles=legendPatches, title="Comment Counts")

axis = plt.gca()
axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)
axis.set_facecolor("whitesmoke")

for i in range(7):
    if i == 1: continue
    count = len(dfUniqueUser[dfUniqueUser["userLevel"] == i])
    plt.text(i-0.05*(len(str(count))+1), count+75, count)

plt.savefig("User Level Distribution Bar Chart.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="userLevel", data=dfUniqueUser, order=[0,1,2,3,4,5,6], color="tab:red", edgecolor="white", linewidth=0.75, width=1)
sns.countplot(x="userLevel", data=dfUniqueUser[dfUniqueUser["userVip"] != 2], order=[0,1,2,3,4,5,6], color="tab:orange", edgecolor="white", linewidth=0.75, width=1)
sns.countplot(x="userLevel", data=dfUniqueUser[dfUniqueUser["userVip"] == 0], order=[0,1,2,3,4,5,6], color="tab:blue", edgecolor="white", linewidth=0.75, width=1)
plt.title("BA 用户VIP等级分布", fontproperties=cf)
plt.xlabel("用户等级", fontproperties=cf)
plt.ylabel("频率", fontproperties=cf)
plt.margins(x=0.05, y=0.1)
legendPatches = [
    Patch(color="tab:red", label="2"), 
    Patch(color="tab:orange", label="1"), 
    Patch(color="tab:blue", label="0")
]
plt.legend(handles=legendPatches, title="VIP Level")

axis = plt.gca()
axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)
axis.set_facecolor("whitesmoke")

for i in range(7):
    if i == 1: continue
    count = len(dfUniqueUser[dfUniqueUser["userLevel"] == i])
    plt.text(i-0.05*(len(str(count))+1), count+75, count)

plt.savefig("User Level & VIP Distribution Bar Chart.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
countBuckets = [0, 1, 2, 3, 4, 5, float('inf')]
countBucketLabels = ['1', '2', '3', '4', '5', '>5']

counts = pd.cut(dfUniqueUser["frequency"], bins=countBuckets, labels=countBucketLabels)
counts = counts.value_counts().sort_index()

counts
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(dfUniqueUser["frequency"].value_counts())
sns.scatterplot(dfUniqueUser["frequency"].value_counts(),s=25, linewidth=0)

plt.title("BA 用户发送评论数分布", fontproperties=cf)
plt.xlabel("用户发送评论条数", fontproperties=cf)
plt.ylabel("人数", fontproperties=cf)
plt.margins(x=0.05, y=0.1)
plt.ylim(-200, 4200)

axis = plt.gca()
axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)
axis.set_facecolor("whitesmoke")
plt.savefig("User Number of Comments Distribution.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
temp = dfUniqueUser["frequency"].value_counts()
temp = temp.reset_index()
temp.columns = ["sent", "number"]
temp = temp.sort_values("sent")
temp["total"] = temp["sent"] * temp["number"]
temp["cumulativeTotal"] = temp["total"].cumsum()
temp
```

```python
fig, ax1 = plt.subplots(figsize=(6, 4))

sns.lineplot(dfUniqueUser["frequency"].value_counts(), ax=ax1, color="tab:blue", label="人数", legend=False)
sns.scatterplot(dfUniqueUser["frequency"].value_counts(),s=25, linewidth=0, ax=ax1, color="tab:blue", legend=False)
ax1.set_xlabel("用户发送评论条数", fontproperties=cf)
ax1.set_ylabel("人数", fontproperties=cf)

ax2 = ax1.twinx()
sns.lineplot(temp, x="sent", y="cumulativeTotal", ax=ax2, color="tab:orange", label="总数")
ax2.set_ylabel("累积评论总数", fontproperties=cf)

plt.xlabel("用户发送评论条数", fontproperties=cf)
plt.margins(x=0.05, y=0.1)

ax1.set_ylim((-200, 4200))
ax2.set_ylim((-10000, 280000))

ax1.spines['top'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax1.set_facecolor("whitesmoke")

line1, label1 = ax1.get_legend_handles_labels()
line2, label2 = ax2.get_legend_handles_labels()
ax2.legend(line1+line2, label1+label2, loc="center right", prop=cf)

plt.savefig("User Number of Comments Distribution & Cumulative Sum.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
dfTime = dataFrame.copy()
dfTime["timeProcessed"] = pd.to_datetime(dfTime["time"], unit="s")
frequencyPerHour = dfTime.groupby(pd.Grouper(key="timeProcessed", freq="h")).size()

plt.figure(figsize=(24, 4))
sns.lineplot(data=frequencyPerHour, legend=False)
axis = plt.gca()
axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)
axis.set_facecolor("whitesmoke")
plt.grid(True, which="both", color="white", linestyle="-", linewidth=1.5)

plt.title("BA 每小时发送评论数趋势", fontproperties=cf)
plt.xlabel("发送时间（小时）", fontproperties=cf)
plt.ylabel("条数", fontproperties=cf)
# plt.savefig("Number of Comments Per Hour.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
cumulativeFrequency = frequencyPerHour.cumsum()
plt.figure(figsize=(16, 7))
sns.lineplot(data=cumulativeFrequency, legend=False)
axis = plt.gca()
axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)
axis.set_facecolor("whitesmoke")
plt.grid(True, which="both", color="white", linestyle="-", linewidth=1.5)

plt.title("BA 评论总数趋势", fontproperties=cf)
plt.xlabel("发送时间", fontproperties=cf)
plt.ylabel("总条数", fontproperties=cf)
plt.savefig("Total Number of Comments.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
dfNewUser = dataFrame.copy()
dfNewUser["timeProcessed"] = pd.to_datetime(dfNewUser["time"], unit="s")
dfNewUser = dfNewUser.sort_values("timeProcessed")

dfNewUser = dfNewUser.drop_duplicates(subset="userId", keep="first")
dfNewUser.set_index("timeProcessed", inplace=True)
dfNewUser = dfNewUser.resample("h").size()
```

```python
plt.figure(figsize=(24, 4))
sns.lineplot(data=dfNewUser, legend=False)
# sns.lineplot(data=frequencyPerHour, legend=False)
axis = plt.gca()
axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)
axis.set_facecolor("whitesmoke")
plt.grid(True, which="both", color="white", linestyle="-", linewidth=1.5)

plt.title("BA 评论区新参与用户数趋势", fontproperties=cf)
plt.xlabel("加入评论区时间", fontproperties=cf)
plt.ylabel("人数", fontproperties=cf)
plt.savefig("Number of New Participants.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(24, 4))

sns.lineplot(data=dfNewUser, ax=ax1, color="tab:blue", label="评论区新参与用户数", legend=False)
ax1.set_xlabel("时间", fontproperties=cf)
ax1.set_ylabel("人数", fontproperties=cf)

ax2 = ax1.twinx()
sns.lineplot(data=frequencyPerHour, ax=ax2, color="tab:orange", label="每小时评论数")
ax2.set_ylabel("条数", fontproperties=cf)

plt.margins(x=0.05, y=0.1)

# ax1.set_ylim((-20, 400))
ax1.set_yticks([0, 50, 100, 150, 200, 250, 300])
# ax2.set_ylim((-200, 9000))
ax2.set_yticks([0, 2000, 4000, 6000, 8000])

ax1.spines['top'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax1.set_facecolor("whitesmoke")

line1, label1 = ax1.get_legend_handles_labels()
line2, label2 = ax2.get_legend_handles_labels()
ax2.legend(line1+line2, label1+label2, loc="center right", prop=cf)

plt.savefig("Comparison of New Participants and Hourly Comments.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
dfTime = dataFrame.copy()
dfTime["timeProcessed"] = pd.to_datetime(dfTime["time"], unit="s")
frequencyPerHourVip0 = dfTime[dfTime["userVip"]==0].groupby(pd.Grouper(key="timeProcessed", freq="h")).size()
frequencyPerHourVip1 = dfTime[dfTime["userVip"]==1].groupby(pd.Grouper(key="timeProcessed", freq="h")).size()
frequencyPerHourVip2 = dfTime[dfTime["userVip"]==2].groupby(pd.Grouper(key="timeProcessed", freq="h")).size()

plt.figure(figsize=(24, 4))
sns.lineplot(data=frequencyPerHourVip2, color="tab:red", label="至少年度大会员用户评论数")
sns.lineplot(data=frequencyPerHourVip1, color="tab:orange", label="月会员用户评论数")
sns.lineplot(data=frequencyPerHourVip0, color="tab:blue", label="免费用户评论数")
plt.legend(prop=cf)
plt.title("BA 根据用户VIP等级划分的评论数量趋势", fontproperties=cf)
plt.xlabel("时间", fontproperties=cf)
plt.ylabel("每小时评论数", fontproperties=cf)
axis = plt.gca()
axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)
axis.set_facecolor("whitesmoke")
plt.grid(True, which="both", color="white", linestyle="-", linewidth=1.5)

plt.savefig("Number of Comments Per Hour By VIP Level.png", dpi=600, bbox_inches='tight')
plt.show()
```
